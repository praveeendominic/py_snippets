```python
#CARS_RAW DATASET
import pandas as pd
import numpy as np

labels = ['mpg',
 'cylinders',
 'displacement',
 'horsepower',
 'weight',
 'acceleration',
 'model year',
 'origin',
 'name']

df = pd.read_csv("cars_raw.csv", skiprows=2,skipfooter=1, names=labels)
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.head()
```

```python
df.rename(mapper={"model year":"model_year"}, axis=1, inplace=True) #Rename specific columns
df.head()
```

```python
df.set_index('name',inplace=True) #Set index
df.head()
```

```python
df['tmp'] = 0 #Add Column at the end,gets broadcasted for the newly added column
df.head()
```

```python
# Add column at specific location
df.insert(loc=1, column='tmp1',value=0)#Inserts a column in index 1
df.head()
```

```python
#Lets now delete columns tmp1 and tmp

df.drop(['tmp1','tmp'], axis=1, inplace=True)
df.head()
```

```python
df.reset_index(inplace=True) #drops the index
df.head()
```

```python
#Append a row
df = pd.concat([df,pd.DataFrame(data=[['TOYOTA',22,22,2222,"2222 hp",222,22,2022,'Canada']], columns=df.columns)], axis=0, ignore_index=True) 
df.tail()
```

```python
df.loc[df.name=='TOYOTA'] #read the row that we have appended
```

```python
#delete the row that we have appended
df.drop(329, axis=0, inplace=True) #since df is range indexed
df.tail()
```

```python
# Extract all cars manufactured in europe with cylinders less than 5
df.loc[((df.origin=='europe') & (df.cylinders<5)),:]
```
